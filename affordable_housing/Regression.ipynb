```python
from pandas import DataFrame as df
```

```python
charges=df.from_csv('charges_month.csv',index_col=None)
```

```python
x=df.from_csv('sell_price.csv',index_col=None)
```

```python
x=x[x.zip_code>1]
```

```python
import pandas as pd
full_table=pd.merge(x,charges,on=['zip_code','year','month'])
```

```python
full_table.dropna
```

```python
from pandas.stats.api import ols
res = ols(y=full_table['count'], x=full_table.iloc[:,:-1])
```
